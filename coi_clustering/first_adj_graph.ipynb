```python
import networkx as nx
import json
import matplotlib.pyplot as plt
```

```python
js = json.load(open('/missouri.json'))
g = nx.readwrite.json_graph.adjacency_graph(js)
```

```python
pos = dict()
for n,d in g.nodes(data=True):
  pos[n] = (float(d['INTPTLON10']),float(d['INTPTLAT10']))
```

```python
nx.draw(g, pos = pos, node_size=1)
```
