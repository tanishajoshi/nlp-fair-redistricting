```python
import pandas as pd
import evaltools.data as ed
import us
```

```python
# key for sorting clusters
cluster_sort_key = lambda x: x.str.split('-').map(lambda y: list(map(int, y)))
```

```python
# read submissions
df_submissions = pd.read_csv('submission_db.csv', usecols=['blocks_2020', 'clusters'], converters={'blocks_2020': eval})
df_clusters = df_submissions.explode('blocks_2020').groupby('clusters').agg(pd.unique).sort_index(key=cluster_sort_key)
```

```python
df_clusters
```

```python
df_pop = ed.census(us.states.MO)
```

```python
# merge population info
df = df_clusters.reset_index().explode('blocks_2020').merge(df_pop, left_on='blocks_2020', right_on='GEOID20', how='left')
```

```python
# population totals by cluster
df_summary = df.groupby('clusters').sum(numeric_only=True).sort_index(key=cluster_sort_key)
```

```python
df_summary
```

```python
df_summary.to_csv('mo_cluster_demographics_p1.csv')
```
